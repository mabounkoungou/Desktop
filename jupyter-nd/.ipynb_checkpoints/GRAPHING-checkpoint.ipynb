```python
import matplotlib.pyplot
import random
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [random.randint(1,200) for _ in range(len(categories))]
plt.figure(figsize = (8,8))
explode = (0.1,0,0,0,0)
plt.pie(values, labels = categories, explode = explode, autopct ="%1.1f%%", shadow = True, startangle = 120)
plt.title("Random Data Pie Chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = 1
b = 2
c = -3
x = np.linspace(-10, 10, 100)
y = a * x**2 + b * x + c
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Graph: y = ax^2 + bx + c')
 plt.grid(False)
plt.axhline(y=0, color='k', linewidth=0.5)  # Add x-axis
plt.axvline(x=0, color='k', linewidth=0.5)  # Add y-axis
plt.show()
```

```python
x,y = 24,36
gcd = 1
for i in range(1, min(x,y)+1):
    if x%i == 0 and y%i == 0:
        gcd = i
print(gcd)
```
